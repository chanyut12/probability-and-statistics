```python
#1.จงหา  P(F8,12<6.79)  และจงสร้างกราฟการแจงแจกแบบเอฟ
from scipy.stats import f
import numpy as np
import matplotlib.pyplot as plt

df1 = 8
df2 = 12
k = 6.79

p =f.cdf(k,df1,df2)
print('P(F8,12<6.79) =',p)
fig, ax = plt.subplots()

# for distribution curve
x= np.arange(0,15,0.001)
ax.plot(x, f.pdf(x,df1,df2))
ax.set_title("F distribution")
ax.set_xlabel('x')
ax.grid(True)

# for fill_between
px=np.arange(0,k,0.01)
ax.fill_between(px,f.pdf(px,df1,df2),alpha=0.4, color='g')

# for text
ax.text(0.5,0.025,round(p,6), fontsize=20)
plt.rcParams["figure.figsize"] = (12, 8)
plt.show()
```

```python
#2 . จงหา  P(1.86<F9,10<3.91)  และจงสร้างกราฟการแจงแจกแบบเอฟ
from scipy.stats import f
import numpy as np
import matplotlib.pyplot as plt

df1 = 9
df2 = 10
k1 = 1.89
k2 = 3.91

p1 =f.cdf(k1,df1,df2)
p2 =f.cdf(k2,df1,df2)
p3 = p2-p1
print('P(1.86<F9,10<3.91) =',p3)
fig, ax = plt.subplots()

# for distribution curve
x= np.arange(0,15,0.001)
ax.plot(x, f.pdf(x,df1,df2))
ax.set_title("F distribution")
ax.set_xlabel('x')
ax.grid(True)

# for fill_between
px=np.arange(k1,k2,0.01)
ax.fill_between(px,f.pdf(px,df1,df2),alpha=0.4, color='g')

# for text
ax.text(2,0.025,round(p,6), fontsize=20)
plt.rcParams["figure.figsize"] = (12, 8)
plt.show()
```

```python
#3 จงหาค่า  a  ที่ทำให้  P(F18,6>a)=0.88  และจงสร้างกราฟการแจงแจกแบบเอฟ
from scipy.stats import f
import numpy as np
import matplotlib.pyplot as plt
df1 = 18
df2 = 6
k = f.isf(0.88,df1,df2)
print("a  ที่ทำให้  P(F18,6>a)=0.88 คือ" , k)
p =f.sf(k,df1,df2)
print('P(F>a) =',p)
fig, ax = plt.subplots()
# for distribution curve
x= np.arange(0,10,0.001)
ax.plot(x, f.pdf(x,df1,df2))
ax.set_title("F distribution")
ax.set_xlabel('x')
ax.grid(True)
# for fill_between
px=np.arange(k,10,0.01)
ax.fill_between(px,f.pdf(px,df1,df2),alpha=0.4, color='g')
# for text
ax.text(1,0.025,round(p,6), fontsize=20)
plt.rcParams["figure.figsize"] = (12, 8)
plt.show()
```

```python
# จงหาค่า  a  ที่ทำให้  P(F7,13<a)=0.55  และจงสร้างกราฟการแจงแจกแบบเอฟ
from scipy.stats import f
import numpy as np
import matplotlib.pyplot as plt
df1 = 7
df2 = 13
k = f.ppf(0.55,df1,df2)
print("a  ที่ทำให้  P(F18,6>a)=0.88 คือ" , k)
p =f.cdf(k,df1,df2)
print('P(F>a) =',p)
fig, ax = plt.subplots()
# for distribution curve
x= np.arange(0,10,0.001)
ax.plot(x, f.pdf(x,df1,df2))
ax.set_title("F distribution")
ax.set_xlabel('x')
ax.grid(True)
# for fill_between
px=np.arange(k,10,0.01)
ax.fill_between(px,f.pdf(px,df1,df2),alpha=0.4, color='g')
# for text
ax.text(2,0.025,round(p,6), fontsize=20)
plt.rcParams["figure.figsize"] = (12, 8)
plt.show()
```

```python
print('mean of G1 = ',df_1['score'].mean(),', standard deviation of G1 =',df_1['score'].std())
print('mean of G2 = ',df_2['score'].mean(),', standard deviation of G2 =',df_2['score'].std())
print('-----------')
print('cv of G1 = ',df_1['score'].std()/df_1['score'].mean()*100)
print('cv of G2 = ',df_2['score'].std()/df_2['score'].mean()*100)
print('-----------')
q3_g1 = df_1['score'].quantile(0.75)
q1_g1 = df_1['score'].quantile(0.25)
q3_g2 = df_2['score'].quantile(0.75)
q1_g2 = df_2['score'].quantile(0.25)
print('CD of G1 = ',(q3_g1-q1_g1)/(q3_g1+q1_g1))
print('CD of G2 = ',(q3_g2-q1_g2)/(q3_g2+q1_g2))
print('-----------')
fig, ax = plt.subplots()
df_1['score'].plot(kind='density', ax=ax,
figsize=(10,4),title="Score of Midterm",legend=True)


df_2['score'].plot(kind='density' , ax=ax,figsize=(10,4),legend=True)

ax.set_xlabel("Score")
ax.set_ylabel("Probability")
ax.grid(axis='y')
ax.legend(['G1', 'G2'])
ax.axvline(x=df_1['score'].mean(), ymax=0.8,color='blue',alpha=0.5,linestyle='dashed')
ax.axvline(x=df_2['score'].mean(), ymax=0.95,color='orange',alpha=0.5,linestyle='dashed')
```
