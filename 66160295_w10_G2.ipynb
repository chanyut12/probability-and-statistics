```python
# 1.‡∏à‡∏á‡∏´‡∏≤‡∏Ñ‡πà‡∏≤‡∏Ç‡∏≠‡∏á  P(œá2_24‚â§12.875)  ‡πÅ‡∏•‡∏∞‡∏à‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Å‡∏£‡∏≤‡∏ü‡∏Å‡∏≤‡∏£‡πÅ‡∏à‡∏á‡πÅ‡∏à‡∏Å‡πÅ‡∏ö‡∏ö‡πÑ‡∏Ñ‡∏™‡πÅ‡∏Ñ‡∏ß‡∏£‡πå
from scipy.stats import chi2
import numpy as np
import matplotlib.pyplot as plt

k = 12.875
df = 24
p =chi2.cdf(k,df)
print('1.‡∏Ñ‡πà‡∏≤‡∏Ç‡∏≠‡∏á  P(œá2_24‚â§12.875) =',p)
fig, ax = plt.subplots()

# for distribution curve
x= np.arange(0,60,0.001)
ax.plot(x, chi2.pdf(x,df))
ax.set_title("$X^2$(X;0,$1^2$)")
ax.set_xlabel('x')
ax.grid(True)

# for fill_between
px=np.arange(0,k,0.01)
ax.fill_between(px,chi2.pdf(px,df),alpha=0.5, color='g')

# for text
ax.text(11,0.005,round(p,6), fontsize=20)

plt.show()
```

```python
# 2.‡∏à‡∏á‡∏´‡∏≤‡∏Ñ‡πà‡∏≤‡∏Ç‡∏≠‡∏á  P(œá2_15>7.459)  ‡πÅ‡∏•‡∏∞‡∏à‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Å‡∏£‡∏≤‡∏ü‡∏Å‡∏≤‡∏£‡πÅ‡∏à‡∏á‡πÅ‡∏à‡∏Å‡πÅ‡∏ö‡∏ö‡πÑ‡∏Ñ‡∏™‡πÅ‡∏Ñ‡∏ß‡∏£‡πå
from scipy.stats import chi2
import numpy as np
import matplotlib.pyplot as plt

k = 7.459
df = 15
p =chi2.sf(k,df)
print('2.‡∏Ñ‡πà‡∏≤‡∏Ç‡∏≠‡∏á  P(œá2_15>7.459) =',p)
fig, ax = plt.subplots()

# for distribution curve
x= np.arange(0,40,0.001)
ax.plot(x, chi2.pdf(x,df))
ax.set_title("$X^2$(X;0,$1^2$)")
ax.set_xlabel('x')
ax.grid(True)

# for fill_between
px=np.arange(k,40,0.01)
ax.fill_between(px,chi2.pdf(px,df),alpha=0.5, color='g')

# for text
ax.text(12,0.009,round(p,6), fontsize=20)

plt.show()
```

```python
# 3.‡∏à‡∏á‡∏´‡∏≤‡∏Ñ‡πà‡∏≤  a  ‡∏ó‡∏µ‡πà‡∏ó‡∏≥‡πÉ‡∏´‡πâ  œá2_0.7504,9=a  ‡πÅ‡∏•‡∏∞‡∏à‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Å‡∏£‡∏≤‡∏ü‡∏Å‡∏≤‡∏£‡πÅ‡∏à‡∏á‡πÅ‡∏à‡∏Å‡πÅ‡∏ö‡∏ö‡πÑ‡∏Ñ‡∏™‡πÅ‡∏Ñ‡∏ß‡∏£‡πå
#‡∏´‡∏≤‡∏Ñ‡πà‡∏≤ a ‡πÇ‡∏î‡∏¢‡∏ó‡∏µ‡πà P(œá2_9<a) = 0.7504
df = 9
k = chi2.ppf(0.7504,df) # k ‡∏Ñ‡∏∑‡∏≠ a
print('3.‡∏Ñ‡πà‡∏≤  a  ‡∏ó‡∏µ‡πà‡∏ó‡∏≥‡πÉ‡∏´‡πâ  œá2_0.7504,9=a ‡∏Ñ‡∏∑‡∏≠ a =',k)
p =chi2.cdf(k,df)
print('P(x^2<a) =',p)
fig, ax = plt.subplots()

# for distribution curve
x= np.arange(0,25,0.001)
ax.plot(x, chi2.pdf(x,df))
ax.set_title("$X^2$(X;0,$1^2$)")
ax.set_xlabel('x')
ax.grid(True)

# for fill_between
px=np.arange(0,k,0.01)
ax.fill_between(px,chi2.pdf(px,df),alpha=0.4, color='g')

# for text
ax.text(3,0.025,round(p,6), fontsize=20)

plt.show()
```

```python
# 4.‡∏à‡∏á‡∏´‡∏≤‡∏Ñ‡πà‡∏≤  a  ‡∏ó‡∏µ‡πà‡∏ó‡∏≥‡πÉ‡∏´‡πâ  P(œá2_14>a)= 0.2588  ‡πÅ‡∏•‡∏∞‡∏à‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Å‡∏£‡∏≤‡∏ü‡∏Å‡∏≤‡∏£‡πÅ‡∏à‡∏á‡πÅ‡∏à‡∏Å‡πÅ‡∏ö‡∏ö‡πÑ‡∏Ñ‡∏™‡πÅ‡∏Ñ‡∏ß‡∏£‡πå
#‡∏´‡∏≤‡∏Ñ‡πà‡∏≤ a ‡πÇ‡∏î‡∏¢‡∏ó‡∏µ‡πà P(œá2_14>a) = 0.2588
df = 14
k = chi2.isf(0.2588,df) # k ‡∏Ñ‡∏∑‡∏≠ a
print('4.‡∏Ñ‡πà‡∏≤  a  ‡∏ó‡∏µ‡πà‡∏ó‡∏≥‡πÉ‡∏´‡πâ  P(œá2_14>a)= 0.2588 ‡∏Ñ‡∏∑‡∏≠ a =',k)
p =chi2.sf(k,df)
print('P(x^2>a) =',p)
fig, ax = plt.subplots()

# for distribution curve
x= np.arange(0,40,0.001)
ax.plot(x, chi2.pdf(x,df))
ax.set_title("$X^2$(X;0,$1^2$)")
ax.set_xlabel('x')
ax.grid(True)

# for fill_between
px=np.arange(k,40,0.01)
ax.fill_between(px,chi2.pdf(px,df),alpha=0.4, color='g')

# for text
ax.text(18,0.003,round(p,6), fontsize=20)

plt.show()
```

```python
# 5.‡∏à‡∏á‡∏´‡∏≤‡∏Ñ‡πà‡∏≤  a  ‡πÅ‡∏•‡∏∞  b  ‡∏ó‡∏µ‡πà‡∏ó‡∏≥‡πÉ‡∏´‡πâ  P(a‚â§œá2_12‚â§b)=0.60  ‡πÅ‡∏•‡∏∞  P(œá2_12‚â§a)=0.10  ‡πÅ‡∏•‡∏∞‡∏à‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Å‡∏£‡∏≤‡∏ü‡∏Å‡∏≤‡∏£‡πÅ‡∏à‡∏á‡πÅ‡∏à‡∏Å‡πÅ‡∏ö‡∏ö‡πÑ‡∏Ñ‡∏™‡πÅ‡∏Ñ‡∏ß‡∏£‡πå
df = 12
k1 = chi2.ppf(0.1,df) # k1 ‡∏Ñ‡∏∑‡∏≠ a
print('5.‡∏Ñ‡πà‡∏≤  a  ‡∏ó‡∏µ‡πà‡∏ó‡∏≥‡πÉ‡∏´‡πâ  P(œá2_12‚â§a)=0.10 ‡∏Ñ‡∏∑‡∏≠ a =',k1)
k2 = chi2.ppf(0.1+0.6,df) # k2 ‡∏Ñ‡∏∑‡∏≠ b
print('5.‡∏Ñ‡πà‡∏≤  b  ‡∏ó‡∏µ‡πà‡∏ó‡∏≥‡πÉ‡∏´‡πâ  P(a‚â§œá2_12‚â§b)=0.60 ‡πÅ‡∏•‡∏∞ P(œá2_12‚â§a)=0.10 ‡∏Ñ‡∏∑‡∏≠ b =',k2)
p =chi2.cdf(k2,df) - chi2.cdf(k1,df)
print('P(a‚â§œá2_12‚â§b) =',p)
fig, ax = plt.subplots()

# for distribution curve
x= np.arange(0,40,0.001)
ax.plot(x, chi2.pdf(x,df))
ax.set_title("$X^2$(X;0,$1^2$)")
ax.set_xlabel('x')
ax.grid(True)

# for fill_between
px=np.arange(k1,k2,0.01)
ax.fill_between(px,chi2.pdf(px,df),alpha=0.4, color='g')

# for text
ax.text(8,0.025,round(p,6), fontsize=20)

plt.show()
```

```python
# 6.‡∏à‡∏á‡∏´‡∏≤‡∏Ñ‡πà‡∏≤‡∏Ç‡∏≠‡∏á  P(T10<2.179)  ‡πÅ‡∏•‡∏∞‡∏à‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Å‡∏£‡∏≤‡∏ü‡∏Å‡∏≤‡∏£‡πÅ‡∏à‡∏á‡πÅ‡∏à‡∏Å‡πÅ‡∏ö‡∏ö‡∏ó‡∏µ
from scipy.stats import t
import numpy as np
import matplotlib.pyplot as plt

df = 10
k = 2.179
p =t.cdf(k,df)
print('6.‡∏Ñ‡πà‡∏≤‡∏Ç‡∏≠‡∏á P(T10<2.179) =',p)
fig, ax = plt.subplots()

# for distribution curve
x= np.arange(-5,5,0.001)
ax.plot(x, t.pdf(x,df))
ax.set_title("Student-t distribution")
ax.set_xlabel('T')
ax.grid(True)

# for fill_between
px=np.arange(-5,k,0.01)
ax.fill_between(px,t.pdf(px,df),alpha=0.4, color='g')

# for text
ax.text(-1,0.06,round(p,6), fontsize=20)

plt.show()
```

```python
# 7.‡∏à‡∏á‡∏´‡∏≤‡∏Ñ‡πà‡∏≤‡∏Ç‡∏≠‡∏á  P(T10>2.179)  ‡πÅ‡∏•‡∏∞‡∏à‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Å‡∏£‡∏≤‡∏ü‡∏Å‡∏≤‡∏£‡πÅ‡∏à‡∏á‡πÅ‡∏à‡∏Å‡πÅ‡∏ö‡∏ö‡∏ó‡∏µ
df = 10
k = 2.179
p =t.sf(k,df)
print('7.‡∏Ñ‡πà‡∏≤‡∏Ç‡∏≠‡∏á P(T10>2.179) =',p)
fig, ax = plt.subplots()

# for distribution curve
x= np.arange(-5,5,0.001)
ax.plot(x, t.pdf(x,df))
ax.set_title("Student-t distribution")
ax.set_xlabel('T')
ax.grid(True)

# for fill_between
px=np.arange(k,5,0.01)
ax.fill_between(px,t.pdf(px,df),alpha=0.4, color='g')

# for text
ax.text(2.5,0.015,round(p,6), fontsize=20)

plt.show()
```

```python
# 8.‡∏à‡∏á‡∏´‡∏≤‡∏Ñ‡πà‡∏≤‡∏Ç‡∏≠‡∏á  P(‚àí0.802<T20<1.037)  ‡πÅ‡∏•‡∏∞‡∏à‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Å‡∏£‡∏≤‡∏ü‡∏Å‡∏≤‡∏£‡πÅ‡∏à‡∏á‡πÅ‡∏à‡∏Å‡πÅ‡∏ö‡∏ö‡∏ó‡∏µ
df = 20
k1 = -0.802
k2 = 1.037
p =t.cdf(k2,df)-t.cdf(k1,df)
print('8.‡∏Ñ‡πà‡∏≤‡∏Ç‡∏≠‡∏á P(‚àí0.802<T20<1.037) =',p)
fig, ax = plt.subplots()

# for distribution curve
x= np.arange(-5,5,0.001)
ax.plot(x, t.pdf(x,df))
ax.set_title("Student-t distribution")
ax.set_xlabel('T')
ax.grid(True)

# for fill_between
px=np.arange(k1,k2,0.01)
ax.fill_between(px,t.pdf(px,df),alpha=0.4, color='g')

# for text
ax.text(-0.8,0.02,round(p,6), fontsize=20)

plt.show()
```
