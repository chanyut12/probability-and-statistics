```python
#1
import pandas as pd
from pydataset import data
df2 = data("ToothGrowth")
print(df2.head)
df2.info()
```

มีทั้งหมด 60 rows 3 column ได้แก่ len supply dose

```python
#2
rangelen = df2['len'].max() -df2['len'].min()
print('พิสัย',rangelen)
print('ความแปรปรวน',df2['len'].var())
print('ส่วนเบี่ยงเบนมาตรฐาน', df2['len'].std())
```

```python
#3
#4 จงแสดงการกระจายของข้อมูลความยาวของเซลล์ที่มีผลกับการงอกของฟันหนูตะเภา โดยใช้ค่าพิสัยควอไทล์ และนำเสนอข้อมูลในรูปของกราฟ box plot

min_s = df2['len'].min()
max_s = df2['len'].max()
q1 = df2['len'].quantile(0.25)
q3 = df2['len'].quantile(0.75)
iqr = q3-q1

print('min =',min_s,'Q1 =',q1, 'Q3 =',q3,'IQR =',iqr,'max =',max_s)
print('median =',df2['len'].median())
print('f1=',q1-(1.5*iqr),'f2 =',q3+(1.5*iqr))
ax = df2['len'].plot(kind='box',figsize=(10,4),legend=False,vert=False)

ax.set_xlabel("Lenght")
ax.grid('y')
```

ข้อมูลความยาวเซลล์ไม่มีค่าที่ผิดปกติ ข้อมูลมีการแจกแจงแบบสมมาตร

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame({'test1': [3,5,10,6,8],'test2': [52,50,55,48,51] })
print('cr #1 = ',(df['test1'].max()- df['test1'].min())/(df['test1'].max()+ df['test1'].min()))
print('cr #2 = ',(df['test2'].max()- df['test2'].min())/(df['test2'].max()+ df['test2'].min()))
#print("cr = " , ((df.max() = df.min()) / (df.max() + df.min()))
ax1=plt.subplot(1, 2, 1)
df['test1'].plot(kind='density',figsize=(10,4),legend=True)
ax1.set_xlabel("Score Test1")
ax1.set_ylabel("Probability")
plt.subplots_adjust(wspace=0.3)
ax2=plt.subplot(1, 2, 2)
df['test2'].plot(kind='density',legend=True)
ax2.set_xlabel("Score Test2")
ax2.set_ylabel("Probability")
```
