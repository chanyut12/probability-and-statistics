```python
#1จงสร้างกราฟการแจงแจกแบบปกติ เมื่อกำหนดให้ x มีค่าตั้งแต่ -5 ถึง 5 โดยแต่ละค่าห่างกัน 0.025
#และแสดงตำแหน่งของค่าความน่าจะเป็นของตัวแปรสุ่ม Z = -2.73 ในกราฟ โดยกำหนดค่าเฉลี่ยเท่ากับ 0 ส่วนเบี่ยงเบนมาตรฐานเท่ากับ 1
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
x= np.arange(-5,5,0.0025)
Z = -2.73
print('f(Z=-2.73) =',norm.pdf(Z))
ax.set_title('N(X;0,$1^2$)')
ax.plot(x, norm.pdf(x))
ax.set_ylim(0,0.45) # จา กดัค่าแกนต้งัให้อยใู่ นช่วง 0 ถึง 0.45
ax.axhline(y=norm.pdf(Z),xmax= 0.26,color='r',linestyle='dashed')
#horizontal line
#ax.axvline(x=Z, ymax=0.5 ,color='r',alpha=0.5) # vertical line
xplot = ax.plot([Z], [norm.pdf(Z)], marker='o', markersize=15,color="red") # coordinate point
ax.set_xlabel('Z',fontsize=20) # set x label
plt.show()
```

```python
#2 จงหาความน่าจะเป็นของตัวแปรสุ่ม Z < 0.59 และสร้างกราฟการแจงแจกแบบปกติ หา P(z < 0.59)
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

k = 0.59
p =norm.cdf(x=k, loc=0, scale=1)
print('จงหาความน่าจะเป็นของตัวแปรสุ่ม Z < 0.59 or P(Z<0.59) =',p)
fig, ax = plt.subplots()

# for distribution curve
x= np.arange(-4,4,0.001)
ax.plot(x, norm.pdf(x,loc=0,scale=1))
ax.set_title("N(X;0,$1^2$)")
ax.set_xlabel('Z')
ax.grid(True)

# for fill_between
px=np.arange(-4,k,0.01)

ax.fill_between(px,norm.pdf(px,loc=0,scale=1),alpha=1, color='pink')

# for text
ax.text(-1.35,0.1,round(p,4), fontsize=20)
plt.show()
```

```python
#3.โรงงานแห่งหนึ่งผลิตหลอดไฟที่มีอายุการใช้งานเฉลี่ย 900 ชั่วโมง ส่วนเบี่ยงเบนมาตรฐาน 50 ชั่วโมง ถ้าอายุการใช้งานของหลอดไฟมีการแจกแจงแบบปกติ
#3.1จงหาความน่าจะเป็นที่หลอดไฟจะมีอายุการใช้งานน้อยกว่า 500 ชั่วโมง
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

k = 500
mu = 900
sigma = 50
p =norm.cdf(x=k, loc=mu, scale=sigma)
print('ความน่าจะเป็นที่หลอดไฟจะมีอายุการใช้งานน้อยกว่า 500 ชั่วโมง X < 500 or P(X<500) =' , p)
fig, ax = plt.subplots()

# for distribution curve
x= np.arange(500,1300,10)
ax.plot(x, norm.pdf(x,loc=mu,scale=sigma))
ax.set_title("N(X;0,$1^2$)")
ax.set_xlabel('X')
ax.grid(True)

# for fill_between
px=np.arange(400,k,10)

ax.fill_between(px,norm.pdf(px,loc=mu,scale=sigma),alpha=1, color='pink')

# for text
ax.text(800,0.002,round(p,4), fontsize=20)
plt.show()
```

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

# Given values
k = 500  # Threshold value
mu = 900  # Mean
sigma = 50  # Standard deviation

# Calculating the probability
p = norm.cdf(x=k, loc=mu, scale=sigma)
print('Probability that the lifetime of the bulb is less than 500 hours (P(X < 500)) =', p)

# Creating the plot
fig, ax = plt.subplots(figsize=(8, 4))  # Adjusted the figsize to be within reasonable limits

# Plotting the distribution curve
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
ax.plot(x, norm.pdf(x, loc=mu, scale=sigma))
ax.set_title("Normal Distribution of Bulb Lifetime")
ax.set_xlabel('Lifetime of Bulb (Hours)')
ax.set_ylabel('Probability Density')
ax.grid(True)

# Highlighting the area under the curve for X < 500
px = np.linspace(mu - 4*sigma, k, 1000)
ax.fill_between(px, norm.pdf(px, loc=mu, scale=sigma), alpha=0.5, color='pink')

# Adding text
ax.text(600, 0.005, f'P(X < 500) = {p:.20f}', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#3.1จงหาความน่าจะเป็นที่หลอดไฟจะมีอายุการใช้งานน้อยกว่า 500-->800 ชั่วโมง
k = 800
mu = 900
sigma = 50
p =norm.cdf(x=k, loc=mu, scale=sigma)
print('ความน่าจะเป็นที่หลอดไฟจะมีอายุการใช้งานน้อยกว่า 800 ชั่วโมง X < 800 or P(X<800) =' , p)
fig, ax = plt.subplots()

# for distribution curve
x= np.arange(700,1100,2)
ax.plot(x, norm.pdf(x,loc=mu,scale=sigma))
ax.set_title("N(X;0,$1^2$)")
ax.set_xlabel('X')
ax.grid(True)

# for fill_between
px=np.arange(700,k,2)

ax.fill_between(px,norm.pdf(px,loc=mu,scale=sigma),alpha=1, color='pink')

# for text
ax.text(790,0.002,round(p,4), fontsize=20)
plt.show()
```

```python
#3.2จงหาความน่าจะเป็นที่หลอดไฟจะมีอายุการใช้งานมากกว่า 1050 ชั่วโมง
k = 1050
mu = 900
sigma = 50
p =norm.sf(x=k, loc=mu, scale=sigma)
print('ความน่าจะเป็นที่หลอดไฟจะมีอายุการใช้งานน้อยกว่า 1050 ชั่วโมง X < 1050 or P(X<1050) =' , p)
fig, ax = plt.subplots()

# for distribution curve
x= np.arange(700,1100,2)
ax.plot(x, norm.pdf(x,loc=mu,scale=sigma))
ax.set_title("N(X;0,$1^2$)")
ax.set_xlabel('X')
ax.grid(True)

# for fill_between
px=np.arange(k,1100,2)

ax.fill_between(px,norm.pdf(px,loc=mu,scale=sigma),alpha=1, color='r')

# for text
ax.text(790,0.002,round(p,4), fontsize=20)
plt.show()
```

```python
#3.3

from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
k1 = 800
k2 = 1000
mu = 900
sigma = 50
p =norm.cdf(x=k2, loc=mu, scale=sigma)-norm.cdf(x=k1, loc=mu, scale=sigma)
print('P จงหาความน่าจะเป็นที่หลอดไฟจะมีอายุการใช้งาน 800 ถึง 1000 ชั่วโมง 800 < Z ≤ 1000) =',p)
fig, ax = plt.subplots()

# for distribution curve
x= np.arange(700,1100,0.5)
ax.plot(x, norm.pdf(x,loc=mu,scale=sigma))
ax.set_title("N(X;0,$1^2$)")
ax.set_xlabel('Z')
ax.grid(True)

# for fill_between

px=np.arange(k1,k2,0.5)
ax.fill_between(px,norm.pdf(px,loc=mu,scale=sigma),alpha=0.5, color='g')

# for text
ax.text(850,0.002,round(p,4), fontsize=20)
plt.show()
```

```python
#3.4 ถ้า 8% ของหลอดไฟที่มีอายุการใช้งานต่ำสุดจะเป็นหลอดไฟที่ไม่ได้มาตรฐาน จงหาอายุการใช้งานต่่ำสุดโดยประมาณที่จะเป็นหลอดไฟที่ได้มาตรฐาน
#หา a ที่ทำให้ P(x<a) = 0.08
a = norm.ppf(0.08,loc=mu , scale=sigma)
print("3.4 อายุการใช้งานต่ำสุดโดยประมาณที่จะเป็นหลอดไฟที่ได้มาตรฐาน  =" ,a, "ชั่วโมง")
k = a
mu = 900
sigma = 50
p =norm.sf(x=k, loc=mu, scale=sigma)
print('ความน่าจะเป็นที่หลอดไฟจะมีอายุการใช้งานน้อยกว่า 1050 ชั่วโมง X < 1050 or P(X<1050) =' , p)
fig, ax = plt.subplots()

# for distribution curve
x= np.arange(700,1100,2)
ax.plot(x, norm.pdf(x,loc=mu,scale=sigma))
ax.set_title("N(X;0,$1^2$)")
ax.set_xlabel('X')
ax.grid(True)

# for fill_between
px=np.arange(700,k,0.2)

ax.fill_between(px,norm.pdf(px,loc=mu,scale=sigma),alpha=1, color='r')

# for text
ax.text(790,0.002,round(p,4), fontsize=20)
plt.show()
```

```python
#3.5  ถ้า 12% ของหลอดไฟที่มีอายุการใช้งานสูงสุดจะเป็นหลอดไฟที่ถูกนำไปติดป้ายเป็นหลอดไฟพรีเมี่ยม จงหาอายุการใช้งานต่่ำสุดโดยประมาณที่จะเป็นหลอดไฟที่ถูกนำไปติดป้ายเป็นหลอดไฟพรีเมี่ยม
#หา a ที่ทำให้ P(x>a) = 0.12
a = norm.isf(0.12,loc=mu , scale=sigma)
print("3.4 อายุการใช้งานสูงสุดจะเป็นหลอดไฟที่ถูกนำไปติดป้ายเป็นหลอดไฟพรีเมี่ยม  =" ,a, "ชั่วโมง")
k = a
mu = 900
sigma = 50
p =norm.sf(x=k, loc=mu, scale=sigma)
print('ความน่าจะเป็นที่หลอดไฟจะมีอายุสูงสุดจะเป็นหลอดไฟที่ถูกนำไปติดป้ายเป็นหลอดไฟพรีเมี่ยม =' , p)
fig, ax = plt.subplots()

# for distribution curve
x= np.arange(700,1100,0.5)
ax.plot(x, norm.pdf(x,loc=mu,scale=sigma))
ax.set_title("N(X;0,$1^2$)")
ax.set_xlabel('X')
ax.grid(True)

# for fill_between
px=np.arange(k,1100,0.2)

ax.fill_between(px,norm.pdf(px,loc=mu,scale=sigma),alpha=1, color='r')

# for text
ax.text(790,0.002,round(p,4), fontsize=20)
plt.show()
```
