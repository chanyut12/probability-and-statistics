```python
#1.4 จงทดสอบว่าข้อมูลน้ำหนักไขมันเฉลี่ยนี้มาจากประชากรที่มีการแจกแจงแบบปกติหรือไม่
#H0 : ข้อมูลน้ำหนักไขมันเฉลี่ยนี้มาจากประชากรที่มีการแจกแจงแบบปกติ
#H1 : ข้อมูลน้ำหนักไขมันเฉลี่ยนี้ไม่ได้มาจากประชากรที่มีการแจกแจงแบบปกติ

from scipy.stats import anderson
data=[35.9, 39.1, 27.8 ,34.2, 30.3,25.7 ,33.2,28.0, 18.9, 23.8, 29.9, 36.0, 35.0 ,30.4, 28.9 ,15.9, 35.2 ,24.3 ,19.4, 27.3 ,31.4 ,24.3, 33.4, 14.5, 16.0, 21.1, 36.4, 33.4 ,22.0 ,36.7 ,30.7, 24.7 ,23.1, 28.8, 38.8, 20.2,25.9]
result = anderson(data)
print(result)
print('Statistic: %.3f' % result.statistic)
p = 0
for i in range(len(result.critical_values)):
  sl, cv = result.significance_level[i], result.critical_values[i]
if result.statistic <= result.critical_values[i]:
  print('%.3f: %.3f, data looks normal(accept H0)' %(sl, cv))
else:
  print('%.3f: %.3f, data doesn’t look normal(reject H0)' %(sl,cv))
```

ยอมรับ H0 ดังนั้นข้อมูลน้ำหนักมาจากการแจกแจงปกติ

```python
from scipy.stats import shapiro
# data=[22,21,19,25,24,26,23,21,23,22]
stat,p = shapiro(data)

print('stat = %.3f, p = %.3f\n' %(stat,p))
if p> 0.01: #.โจทย์กำหนด ช่วงความเชื่อมั่น 99%
  print('ยอมรับ H0 คือ probably normal ')
else:
  print(('ปฏิเสธ H) คือ probably not normal '))
```

```python
#เมื่อต้องการประมาณจานวนคร้ังขอบการเต้นของหัวใจโดยเฉลี่ยต่อนาทีของประชากรกลุ่มหนึ่งจึง สุ่มบุคลากรจากประชากรกลุ่มนี้มาจำนวน 14 คน ดังนี้
#ถ้าประชากรมีการแจกแจงแบบปกติ จงตอบคาถาม ต่อไปนี้ 87, 89, 95, 90, 91, 92, 93, 95, 90, 89, 88, 87, 92, 86
#2.1  จงหาช่วงความเชื่อมั่น 99% ของจำนวนครั้งของการเต้นของหัวใจโดยเฉลี่ย with unknown variance
from scipy.stats import t,sem
from statistics import mean
data1 = [87, 89, 95, 90, 91, 92, 93, 95, 90, 89, 88, 87, 92, 86]
confidence_level = 0.99
degrees_freedom = len(data1) - 1
sample_mean = mean(data1)
sample_standard_error = sem(data1)

confidence_interval = t.interval(confidence_level, degrees_freedom,sample_mean, sample_standard_error)
print(confidence_interval)
```

ช่วงความเชื่อมั่น 99 %  ของจำนวนครั้งของการเต้นของหัวใจโดยเฉลี่ยอยู่ระหว่างค่า 87.97758132315163 to  92.59384724827696




```python
#2.1
from scipy.stats import t
from statistics import mean,stdev
from math import sqrt
data1 = [87, 89, 95, 90, 91, 92, 93, 95, 90, 89, 88, 87, 92, 86]
n = len(data1)
x_bar = mean(data1)
sd = stdev(data1)

df = len(data1)-1
cf_level = 0.99
print(x_bar,sd)
t = t.ppf((1-cf_level)/2,df)
print('t =',t)
print('L =',x_bar -(-(t)*(sd/sqrt(n))))
print('U =',x_bar +(-(t)*(sd/sqrt(n))))
```

```python
#2.2 จงหาช่วงความเชื่อมั่น 90% ของจำนวนคร้ังของการเต้นของหัวใจโดยเฉลี่ย
from scipy.stats import t,sem
from statistics import mean
data1 = [87, 89, 95, 90, 91, 92, 93, 95, 90, 89, 88, 87, 92, 86]
confidence_level = 0.90
degrees_freedom = len(data1) - 1
sample_mean = mean(data1)
sample_standard_error = sem(data1)

confidence_interval = t.interval(confidence_level, degrees_freedom,sample_mean, sample_standard_error)
print(confidence_interval)
```

ช่วงความเชื่อมั่น 90 %  ของจำนวนครั้งของการเต้นของหัวใจโดยเฉลี่ยอยู่ระหว่างค่า 88.92875032712378 to 91.6426782443048

```python
#3.สุ่มตัวอย่างหญิงไทยที่สิ้นชีวิตไปในที่ผ่านมา 27 คน พบว่ามีอายุเป็น 85 59 60 98 66 76 62 61 66 82 68 72 76 76 70 73 71 87 69 68 70 70 69 78 75 65 73
#จากข้อมูลดังกล่าวสามารถยืนยันได้หรือไม่ว่าอายุเฉลี่ยที่แท้จริงของหญิงไทยในปัจจุบัน มากกว่า 70 ปี โดยใช้ระดับนัยสําคัญ 0.01
#H0 : อายุเฉลี่ยที่แท้จริงของหญิงไทยในปัจจุบัน น้อยกว่าเท่ากับ 70 ปี    ----  mu <= 70
#H1 : อายุเฉลี่ยที่แท้จริงของหญิงไทยในปัจจุบัน มากกว่า 70 ปี       ---- mu > 70


#test norm or not
from scipy.stats import shapiro

data2=[85 ,59 ,60, 98, 66 ,76, 62 ,61, 66, 82, 68 ,72 ,76, 76, 70, 73 ,71, 87 ,69 ,68, 70, 70, 69, 78 ,75, 65, 73]
stat,p = shapiro(data2)
print('stat = %.3f, p = %.3f\n' %(stat,p))
if p > 0.01:
  print('Probably Normal accepted H0 :')
else:
  print(('Probably not Normal unaccepted H0 : '))
```

```python
#3.สุ่มตัวอย่างหญิงไทยที่สิ้นชีวิตไปในที่ผ่านมา 27 คน พบว่ามีอายุเป็น 85 59 60 98 66 76 62 61 66 82 68 72 76 76 70 73 71 87 69 68 70 70 69 78 75 65 73
#จากข้อมูลดังกล่าวสามารถยืนยันได้หรือไม่ว่าอายุเฉลี่ยที่แท้จริงของหญิงไทยในปัจจุบัน มากกว่า 70 ปี โดยใช้ระดับนัยสําคัญ 0.01
#H0 : อายุเฉลี่ยที่แท้จริงของหญิงไทยในปัจจุบัน น้อยกว่าเท่ากับ 70 ปี    ----  mu <= 70
#H1 : อายุเฉลี่ยที่แท้จริงของหญิงไทยในปัจจุบัน มากกว่า 70 ปี       ---- mu > 70

#test wih t tester
from scipy.stats import ttest_1samp
data2=[85 ,59 ,60, 98, 66 ,76, 62 ,61, 66, 82, 68 ,72 ,76, 76, 70, 73 ,71, 87 ,69 ,68, 70, 70, 69, 78 ,75, 65, 73]

popmean = 70
stat, p = ttest_1samp(data2, popmean, alternative='greater')

print('stat = %.7f, p = %.7f\n' %(stat,p))
if p <= 0.01: # alpha value is 0.01 or 1%
  print(" we are rejecting null hypothesis unaccepted H0 :อายุเฉลี่ยที่แท้จริงของหญิงไทยในปัจจุบัน มากกว่า 70 ปี ")
else:
  print("we are accepting null hypothesis accepted H0 : อายุเฉลี่ยที่แท้จริงของหญิงไทยในปัจจุบัน น้อยกว่าเท่ากับ  70 ปี")
```
